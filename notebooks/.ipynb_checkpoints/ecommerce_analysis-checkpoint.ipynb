```python

import pandas as pd
import sqlalchemy

db_path = '../database/ecommerce.db'

engine = sqlalchemy.create_engine(f'sqlite:///{db_path}')

query = """
SELECT
    c.customer_unique_id,
    o.order_id,
    o.order_status,
    o.order_purchase_timestamp,
    oi.price,
    oi.freight_value
FROM
    customers c
JOIN
    orders o ON c.customer_id = o.customer_id
JOIN
    order_items oi ON o.order_id = oi.order_id
WHERE
    o.order_status = 'delivered';
"""

# Use pandas to execute the query and load the result into a DataFrame
df = pd.read_sql_query(query, engine)

# display first 5 rows
print(df.head())

# displany common info
print(df.info())
```

```python
# check for missing vals
df.isnull().sum() 
# convert timestamp to datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
# total price for each item
df['total_price'] = df['price'] + df['freight_value']
```

```python
# set the timestamp as the index to easily analyze sales over time
df_eda = df.set_index('order_purchase_timestamp')

# resample by month to see monthly sales
monthly_sales = df_eda['total_price'].resample('ME').sum()

# Plot the results
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))
monthly_sales.plot(title='Total Monthly Sales')
plt.show()
```

```python
# find the most recent date in the dataset
snapshot_date = df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# calculate recency for each customer
recency = df.groupby('customer_unique_id').agg(
    last_purchase_date=('order_purchase_timestamp', 'max')
).reset_index()
recency['recency'] = (snapshot_date - recency['last_purchase_date']).dt.days
```

```python
# calculate frequency for each customer
frequency = df.groupby('customer_unique_id').agg(
    frequency=('order_id', 'count')
).reset_index()
```

```python
# calculate total price for each customer
monetary = df.groupby('customer_unique_id').agg(
    monetary=('total_price', 'sum')
).reset_index()
```
